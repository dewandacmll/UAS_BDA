```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data jumlah pembeli per hari dari Januari ke Februari
data_januari = np.random.randint(50, 150, size=31)
data_februari = np.random.randint(100, 200, size=29)
```

```python
# Menggabungkan data
data_pembeli = np.concatenate((data_januari, data_februari))

# Membuat DataFrame
dates = pd.date_range(start='2024-01-01', periods=len(data_pembeli))
df = pd.DataFrame(data={'Jumlah Pembeli': data_pembeli}, index=dates)
```

```python
# Q-Learning parameters
alpha = 0.1  # learning rate
gamma = 0.9  # discount factor
epsilon = 0.1  # exploration rate
n_states = len(data_pembeli)
n_actions = 3  # Increase, Decrease, Stay
```

```python
# Initialize Q-table
Q = np.zeros((n_states, n_actions))
```

```python
# Reward function: Reward if the number of buyers increases
def get_reward(state, action):
    if action == 0:  # Increase
        return df.iloc[state + 1]['Jumlah Pembeli'] - df.iloc[state]['Jumlah Pembeli']
    elif action == 1:  # Decrease
        return df.iloc[state]['Jumlah Pembeli'] - df.iloc[state + 1]['Jumlah Pembeli']
    else:  # Stay
        return 0
```

```python
# Training Q-Learning algorithm
for episode in range(1000):
    state = 0
    while state < n_states - 1:
        if np.random.rand() < epsilon:
            action = np.random.choice(n_actions)  # Explore
        else:
            action = np.argmax(Q[state])  # Exploit

        reward = get_reward(state, action)
        next_state = state + 1

        Q[state, action] = Q[state, action] + alpha * (reward + gamma * np.max(Q[next_state]) - Q[state, action])

        state = next_state
```

```python
# Best actions to take from each state
best_actions = np.argmax(Q, axis=1)

# Create a new column for best action in the DataFrame
df['Best Action'] = best_actions

# Display the DataFrame
print(df)

# Plotting
plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Jumlah Pembeli'], label='Jumlah Pembeli')
plt.xlabel('Date')
plt.ylabel('Jumlah Pembeli')
plt.title('Grafik Jumlah Pembeli Online Shop')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Fungsi untuk mengecek jumlah pembeli pada tanggal tertentu
def cek_jumlah_pembeli(tanggal):
    try:
        jumlah_pembeli = df.loc[tanggal, 'Jumlah Pembeli']
        print(f"Jumlah pembeli pada tanggal {tanggal} adalah {jumlah_pembeli}.")
    except KeyError:
        print("Tanggal tidak ditemukan dalam data karena data ini hanya memuat data pada bulan januari dan februari, tolong berikan tanggal yang tepat.")
```

```python
# Fungsi untuk membandingkan jumlah pembeli antara bulan Januari dan Februari
def bandingkan_jumlah_pembeli():
    jumlah_januari = df.loc['2024-01-01':'2024-01-31', 'Jumlah Pembeli'].sum()
    jumlah_februari = df.loc['2024-02-01':'2024-02-29', 'Jumlah Pembeli'].sum()
    
    print(f"Jumlah pembeli pada bulan Januari: {jumlah_januari}")
    print(f"Jumlah pembeli pada bulan Februari: {jumlah_februari}")
    
    if jumlah_januari > jumlah_februari:
        print("Jumlah pembeli pada bulan Januari lebih banyak dibandingkan dengan bulan Februari.")
    elif jumlah_januari < jumlah_februari:
        print("Jumlah pembeli pada bulan Februari lebih banyak dibandingkan dengan bulan Januari.")
    else:
        print("Jumlah pembeli pada bulan Januari dan Februari sama.")
```
